```python
# Path locations (folder location) for respective files
dir_paraquet = 'D:/rohan/thesis/Projects/Machine_Learning/Data/Manufacturing_Data/'
dir_planning = 'D:/rohan/thesis/Projects/Machine_Learning/Data/Planning_Data/'
dir_compensation = 'D:/rohan/thesis/Projects/Machine_Learning/Data/Compensation_Data/'
dir_final_save  = 'D:/rohan/thesis/Projects/Machine_Learning/Data/Combined_Data/'
dir_block_save  = 'D:/rohan/thesis/Projects/Machine_Learning/Data/Combined_Data/Blockwise/'
dir_blisk_save  = 'D:/rohan/thesis/Projects/Machine_Learning/Data/Combined_Data/Blisk/'
```

```python
# All blocks and angles for whole blisk
blocks = [1030, 2030, 3030, 4030, 5030, 6030, 7030]
angles = list(np.linspace(0,348,30))
tol = 3
```

### Complete Blisk Data Generation
Loop throgh all the blocks and angles and concatinate block wise

```python
t0 = time.time()
for block in blocks:
    for angle in angles:
        print('The Block {0} and angle {1}'.format(int(block), int(angle)))
        generator(dir_paraquet, dir_planning,dir_compensation, dir_final_save, int(block), int(angle), tol)
        
t1 = time.time()
```

```python
t2 = time.time()
for block in blocks:
    dfs = []
    for angle in angles:
        print(block, angle)
        dfs.append(pd.read_excel(str(dir_final_save)+'finaldf_forward_with_compensation'+str(block)+'__'+str(int(angle))+'.xlsx'))
        df = pd.concat(dfs)
    df.to_csv(str(dir_block_save)+str(block)+'.csv', mode='a', chunksize='10000',index=False)

t3 = time.time()
```

```python
t4 = time.time()
dfss = []
for block in blocks:
    print(block)
    dfss.append(pd.read_csv(str(dir_block_save)+str(block)+'.csv'))

df_blisk = pd.concat(dfss)
df_blisk.to_csv(str(dir_blisk_save)+'full_blisk'+'.csv', mode='a', chunksize='10000',index=False)
```
