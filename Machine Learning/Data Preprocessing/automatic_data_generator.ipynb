```python
block = 3030
```

```python
dfs = []
for angle in angles:
    print(angle)
    dfs.append(pd.read_excel(str(dir_final_save)+'finaldf_forward_with_compensation'+str(block)+'__'+str(int(angle))+'.xlsx'))
```

```python
summ = 0
count = 0
for i in dfs:
    summ = summ + i.shape[0]
    count = count + 1
    print(summ, count)
```

```python
df = pd.concat(dfs)
```
